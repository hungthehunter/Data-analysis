```python
# Practice 2
import random
import math

# 1. Tạo dãy số a gồm 10 phần tử ngẫu nhiên từ 1 đến 100
N = 10
a = [random.randint(1, 100) for _ in range(N)]
print("Dãy số a:", a)

# 2a. Bình phương các phần tử
a_square = [x**2 for x in a]
print("Bình phương các phần tử:", a_square)

# 2b. Tìm số lượng phần tử
count = len(a)
print("Số lượng phần tử:", count)

# 2c. In ra max, min, trung bình
max_val = max(a)
min_val = min(a)
mean_val = sum(a) / count
print("Giá trị lớn nhất:", max_val)
print("Giá trị nhỏ nhất:", min_val)
print("Giá trị trung bình:", mean_val)

# 2d. Tính phương sai và độ lệch chuẩn
sigma2 = sum((x - mean_val) ** 2 for x in a) / (count - 1)   # phương sai
std_dev = math.sqrt(sigma2)                                  # độ lệch chuẩn
print("Phương sai (σ^2):", sigma2)
print("Độ lệch chuẩn (var):", std_dev)
```

```python
# Practice 3
import numpy as np
import random
import math

# Tạo dãy a ngẫu nhiên có 10 phần tử (1..100)
N = 10
a = [random.randint(1, 100) for _ in range(N)]
print("Dãy a:", a)

# 3a. Tạo dãy số x có 10 phần tử từ 1 đến 10
x = np.arange(1, 11)
print("Dãy x:", x)

# 3b. Liệt kê và đếm số lượng phần tử chẵn, lẻ
a_even = [val for val in a if val % 2 == 0]
a_odd = [val for val in a if val % 2 != 0]
x_even = [val for val in x if val % 2 == 0]
x_odd = [val for val in x if val % 2 != 0]

print("\nDãy a - chẵn:", a_even, "số lượng:", len(a_even))
print("Dãy a - lẻ:", a_odd, "số lượng:", len(a_odd))
print("Dãy x - chẵn:", x_even, "số lượng:", len(x_even))
print("Dãy x - lẻ:", x_odd, "số lượng:", len(x_odd))

# 3c. Khoảng cách s giữa các phần tử ở vị trí lẻ
# (chỉ lấy index lẻ: 1, 3, 5,... tức là theo Python index % 2 != 0)
s = math.sqrt(sum((a[i] - x[i])**2 for i in range(N) if i % 2 != 0))
print("\nKhoảng cách s giữa các phần tử ở vị trí lẻ:", s)

# 3d. Tìm khoảng cách nhỏ nhất giữa tập a và x
# Khoảng cách giữa 2 tập = min(|ai - xj|) với mọi ai thuộc a, xj thuộc x
min_dist = min(abs(ai - xj) for ai in a for xj in x)
print("Khoảng cách nhỏ nhất giữa tập a và x:", min_dist)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# (a) Sinh ngẫu nhiên dãy số có 100 phần tử phân bố đều [1, 10]
uniform_data = np.random.uniform(1, 10, 100)

plt.hist(uniform_data, bins=10, edgecolor='black')
plt.title("Phân bố đều Uniform(1,10) - 100 phần tử")
plt.xlabel("Giá trị")
plt.ylabel("Tần số")
plt.show()

# (b) Sinh ngẫu nhiên N điểm (x, y) với y là hàm phân phối chuẩn N(μ=1, σ^2=1)
mu, sigma = 1, 1.0
N = 100
x_vals = np.linspace(mu - 5*sigma, mu + 5*sigma, N)  # N điểm trong khoảng
y_vals = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-((x_vals - mu)**2) / (2*sigma**2))

plt.plot(x_vals, y_vals, 'r-', label="Hàm mật độ chuẩn N(1,1)")
plt.title("Hàm mật độ xác suất chuẩn")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

# (c) Sinh ngẫu nhiên 100000 phần tử theo phân bố chuẩn N(μ=1, σ^2=1)
normal_data = np.random.normal(mu, sigma, 100000)

plt.hist(normal_data, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')
# vẽ thêm đường cong lý thuyết
x_vals = np.linspace(mu - 5*sigma, mu + 5*sigma, 1000)
y_vals = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-((x_vals - mu)**2) / (2*sigma**2))
plt.plot(x_vals, y_vals, 'r-', linewidth=2, label="PDF N(1,1)")

plt.title("Phân bố chuẩn N(1,1) - 100000 phần tử")
plt.xlabel("Giá trị")
plt.ylabel("Mật độ xác suất")
plt.legend()
plt.show()
```

```python
# Practice 2
import numpy as np

# Tạo ma trận a có M=6 dòng, N=4 cột (giá trị ngẫu nhiên từ 1..100)
M, N = 6, 4
a = np.random.randint(1, 101, size=(M, N))
print("Ma trận a:\n", a)

# (a) Ma trận chuyển vị
print("\n(a) Ma trận chuyển vị:\n", a.T)

# (b) Phần tử ở dòng 2 cột 3 (chỉ số Python bắt đầu từ 0 → dòng 2 = index 1, cột 3 = index 2)
print("\n(b) Phần tử dòng 2, cột 3:", a[1, 2])

# (c) Trích xuất dòng đầu, dòng cuối, dòng 2
print("\n(c) Dòng đầu:", a[0])
print("    Dòng cuối:", a[-1])
print("    Dòng 2:", a[1])

# (d) Trích xuất cột kế cuối, cột đầu
print("\n(d) Cột kế cuối:", a[:, -2])
print("    Cột đầu:", a[:, 0])

# (e) Đảo các giá trị trên từng cột
a_flipped = np.flipud(a)
print("\n(e) Ma trận sau khi đảo giá trị trên từng cột:\n", a_flipped)

# (f) Tính tổng các dòng, trung bình các cột
row_sums = np.sum(a, axis=1)
col_means = np.mean(a, axis=0)
print("\n(f) Tổng các dòng:", row_sums)
print("    Trung bình các cột:", col_means)
```

```python
# Practice 3
import numpy as np

# (a) Tạo ma trận A, B (4x3) và X (3x4)
A = np.random.randint(1, 10, size=(4, 3))
B = np.random.randint(1, 10, size=(4, 3))
X = np.random.randint(1, 10, size=(3, 4))

print("Ma trận A (4x3):\n", A)
print("\nMa trận B (4x3):\n", B)
print("\nMa trận X (3x4):\n", X)

# (b) Tổng, hiệu, tích từng phần tử giữa A và B
sum_AB = A + B
diff_AB = A - B
prod_AB = A * B

print("\n(b) Tổng A + B:\n", sum_AB)
print("\n(b) Hiệu A - B:\n", diff_AB)
print("\n(b) Tích từng phần tử A * B:\n", prod_AB)

# (c) Phép nhân ma trận Y = A x X
Y = A @ X   # hoặc np.dot(A, X)
print("\n(c) Ma trận Y = A x X:\n", Y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(N):
    # Sinh ngẫu nhiên N điểm (x,y) trong [-1, 1]
    x = np.random.uniform(-1, 1, N)
    y = np.random.uniform(-1, 1, N)

    # Kiểm tra điểm trong đường tròn
    inside = x**2 + y**2 <= 1
    Nc = np.sum(inside)

    # Ước lượng pi
    pi_approx = 4 * Nc / N
    error = abs(np.pi - pi_approx)
    return pi_approx, error, x, y, inside

# Thử với N = 100, 10000, 1000000
for N in [100, 10000, 1000000]:
    pi_est, err, _, _, _ = monte_carlo_pi(N)
    print(f"N={N:>7}, π xấp xỉ = {pi_est:.6f}, sai số = {err:.6f}")

# Vẽ minh họa với N nhỏ (ví dụ N=500)
pi_est, _, x, y, inside = monte_carlo_pi(500)

plt.figure(figsize=(6,6))
plt.scatter(x[inside], y[inside], color="blue", s=5, label="Trong đường tròn")
plt.scatter(x[~inside], y[~inside], color="red", s=5, label="Ngoài đường tròn")
circle = plt.Circle((0,0), 1, color="green", fill=False, linewidth=2, label="Đường tròn đơn vị")
plt.gca().add_patch(circle)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f"Monte Carlo π ≈ {pi_est:.4f} (N=500)")
plt.legend()
plt.show()
```
